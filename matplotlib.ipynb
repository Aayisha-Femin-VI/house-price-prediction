```python
import matplotlib.pyplot as plt
```

```python

x=(1,2,3,4)
y=(10,30,20,10)
plt.plot(x,y)
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```

```python
x=(1,2,3,4)
y1=(10,20,20,100)
y2=(10,20,30,40)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```

```python
catogories=('A','B','C','D')
value=(10,30,20,50)
plt.bar(catogories,value)
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```

```python
value=[1,2,2,3,3,3,4]
plt.hist(value)
plt.show()
```

```python
catogories=('A','B','C','D')
value=(10,15,25,30)
plt.scatter(catogories,value)
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.show()
```

```python
# piechart
catogories=('A','B','C','D')
value=(10,30,20,50)
plt.pie(value,labels=catogories)
plt.show()
```

```python
#customizable graph
x=(1,2,3,4)
y=(10,30,70,40)
plt.plot(x,y,linestyle='--',marker='o')
plt.grid()
plt.show()
```
